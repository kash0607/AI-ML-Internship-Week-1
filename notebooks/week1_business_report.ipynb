# Task 3: Business Insight Report

# Objective

The objective of this report is to identify factors associated with student academic performance and provide actionable recommendations based on data-driven analysis.

# Dataset Summary

The dataset contains 1000 student records.

The analysis includes demographic variables such as gender, ethnicity, parental education, lunch type, and test preparation status, along with academic performance indicators in Mathematics, Reading, and Writing.

An AverageScore variable was created to represent overall academic performance.

# Analysis Approach

The analysis was performed in three stages:

1. Data inspection and cleaning.
2. Exploratory data analysis and visualization.
3. Business-oriented interpretation of findings.

Statistical summaries, grouping analysis, correlation studies, and visualizations were used to identify patterns within the dataset.

# Executive Summary

This report analyzes student academic performance using demographic and educational factors.

The dataset contains 1000 student records with examination scores in Mathematics, Reading, and Writing, along with supporting demographic information.

The objective of this analysis is to identify patterns that influence academic performance and provide actionable insights based on the findings.

# Business Question 1

Does test preparation influence student performance?

### Finding

Students who completed the test preparation course achieved an average score of approximately 72.67.

Students who did not complete the course achieved an average score of approximately 65.04.

This represents an improvement of approximately 7.6 points.

### Business Insight

Test preparation programs appear to have a positive impact on academic performance.

Educational institutions may improve student outcomes by encouraging participation in preparation courses and providing additional learning resources.

# Business Question 2

Is there a relationship between reading and writing performance?

### Finding

Reading and writing scores exhibit a very strong positive correlation (approximately 0.95).

Students with higher reading scores generally achieve higher writing scores.

### Business Insight

Reading and writing skills appear closely connected.

Programs designed to improve reading comprehension may also contribute to improvements in writing performance.

# Business Question 3

Does parental education influence student outcomes?

### Finding

Students whose parents have higher education levels tend to achieve slightly higher average scores.

Average performance increases gradually from the high-school category to the master's degree category.

### Business Insight

Family educational background may influence student achievement.

Schools may consider providing additional academic support for students who have fewer educational resources available at home.

# Business Question 4

What is the overall academic performance distribution?

### Finding

Most students achieved average scores between 60 and 80 marks.

The average overall score was approximately 67.77.

### Business Insight

The majority of students demonstrate moderate academic performance.

Targeted intervention programs could be developed for lower-performing students to improve overall outcomes.

# Key Findings

1. Students who completed test preparation scored approximately 7.6 points higher on average.

2. Reading and writing scores show a very strong positive correlation.

3. Most students achieved average scores between 60 and 80.

4. Academic performance indicators are strongly correlated.

5. Students with higher parental education levels tend to achieve slightly higher scores.

# Recommendations

Based on the analysis, the following recommendations are proposed:

1. Increase participation in test preparation programs.

2. Strengthen reading-focused educational initiatives to support both reading and writing performance.

3. Provide additional support mechanisms for students who may lack academic assistance outside school.

4. Implement early intervention programs for students demonstrating lower performance levels.

5. Continue monitoring academic performance using data-driven approaches.

# Limitations

1. The dataset does not include attendance information.

2. The dataset does not include study hour information.

3. Correlation does not imply causation.

4. External factors such as socioeconomic conditions and teaching quality are not represented in the dataset.

5. Findings should be interpreted as observed associations rather than direct causal relationships.

# Conclusion

The analysis identified several factors associated with student academic performance.

Test preparation status demonstrated the strongest practical impact on performance outcomes.

Reading and writing skills showed a strong positive relationship, while parental education displayed a moderate influence on achievement levels.

These findings highlight opportunities for educational institutions to improve student outcomes through targeted support programs and evidence-based decision-making.
